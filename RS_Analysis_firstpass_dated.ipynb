```python
#Enter the path for the folder where all the baseline data is stored
datapath_baseline = '/Volumes/PortableSSD/HOI_LLK/RS_Baseline'
```

```python
#Initialize a dictionary for the baseline data
data_baseline_dict = {}
```

```python
#Load files onto a dictionary 
for file_name in os.listdir(datapath_baseline):
    file_path = os.path.join(datapath_baseline, file_name)
    
    # Ignore hidden files (files starting with "._")
    if not file_name.startswith("._") and file_name.endswith('.mat') and len(file_name) >= 10:
        # Extract 8th, 9th, and 10th characters from the filename
        key_name = file_name[7:10]+ "_RS_baseline"
        
        # Load the .mat file using scipy.io.loadmat
        try:
            data_baseline_dict[key_name] = sp.io.loadmat(file_path)
            print(f"Loaded file '{file_name}' into data_baseline_dict['{key_name}']")
        except Exception as e:
            print(f"Error loading file '{file_name}': {e}")
```

```python
#Sort the data to ensure the keys are in order
data_baseline_dict = {key: data_baseline_dict[key] for key in sorted(data_baseline_dict.keys())}
print(data_baseline_dict.keys())
```

```python
def process_data(data_dict):
    """
    Transposes data and applies the multi_order_measures function to each frequency band.
    Stores the processed results in the respective frequency band keys in the processed_data dictionary.

    Parameters:
    - data_dict (dict): A dictionary where each key represents a subject and each value contains raw data with frequency band keys (e.g., 'Alpha', 'Beta').

    Returns:
    - dict: A dictionary where each key is a subject and each value contains processed results for the frequency bands.
    """
    
    results = {}
    
    for key in data_dict.keys():
        theta_data = data_dict[key]['Theta'].transpose()
        alpha_data = data_dict[key]['Alpha'].transpose()
        beta_data = data_dict[key]['Beta'].transpose()
        gamma_data = data_dict[key]['Gamma'].transpose()
        wholeband_data = data_dict[key]['Wholeband'].transpose()
        
        # Process and store results for the current key
        results[key] = {
            'Theta': multi_order_measures(theta_data),
            'Alpha': multi_order_measures(alpha_data),
            'Beta': multi_order_measures(beta_data),
            'Gamma': multi_order_measures(gamma_data),
            'Wholeband': multi_order_measures(wholeband_data) 
        }
    
    return results

# Call the function with your baseline data_dict
processed_data_RS_baseline = process_data(data_baseline_dict)
```

```python
#Check the number of keys in the processed data to ensure correct processing
num_keys = len(processed_data_RS_baseline)
print(f"Number of keys in processed_data_RS_baseline: {num_keys}")

#Then, double check the keys of the processed data
print(processed_data_RS_baseline.keys())

#Check how the multi_order_measure data looks for an example subject
print(processed_data_RS_baseline['004_RS_baseline']['Alpha'])
        
```

### Combining the data and creating separate dictionaries for each subject

```python
import pandas as pd
import numpy as np

def combine_timepoints_data(*data_dicts):
    """
    Combine data from multiple dictionaries into a single dictionary.
    
    Parameters:
    - data_dicts (dict): Dictionaries with processed data for different timepoints.
    
    Returns:
    - dict: Combined data for all timepoints.
    """
    combined_data = {}
    
    for data_dict in data_dicts:
        for key in data_dict:
            if key not in combined_data:
                combined_data[key] = {}
            # Merge frequency band data
            for band in ['Alpha', 'Beta', 'Gamma', 'Theta', 'Wholeband']:
                if band not in combined_data[key]:
                    combined_data[key][band] = {}
                combined_data[key][band].update(data_dict[key][band])
    
    return combined_data

# Combine the data from all timepoints
combined_data = combine_timepoints_data(
    processed_data_RS_baseline,
    processed_data_RS_1H,
    processed_data_RS_24H,
    processed_data_RS_Day7
)
```

```python
#Save combined_data
with open('combined_data.pkl', 'wb') as file:
    pickle.dump(combined_data, file)
```

```python
def create_subject_dictionaries(combined_data):
    """
    Organizes the combined data into separate dictionaries for each subject, 
    ensuring that each subject has entries for all timepoints (with None for missing data).
    
    Parameters:
    - combined_data (dict): The dictionary with all subjects' data for multiple timepoints.

    Returns:
    - dict: A dictionary with each subject ID as a key and another dictionary containing timepoints as values.
    """
    subjects_dict = {}
    timepoints = ['RS_baseline', 'RS_1H', 'RS_24H', 'RS_Day7']
    
    # Organize data by subject
    for key in combined_data.keys():
        # Extract subject ID and timepoint from the key
        subject_id, timepoint = key.split('_RS_')
        
        # Initialize the subject dictionary if not already present
        if subject_id not in subjects_dict:
            subjects_dict[subject_id] = {}
        
        # Add the timepoint data to the subject's dictionary
        subjects_dict[subject_id][f'RS_{timepoint}'] = combined_data[key]
    
    # Ensure each subject has all timepoints (fill missing ones with None)
    for subject_id in subjects_dict:
        for timepoint in timepoints:
            if timepoint not in subjects_dict[subject_id]:
                subjects_dict[subject_id][timepoint] = None

    return subjects_dict

# Example usage:
# Assuming `combined_data` is already a merged dictionary with timepoint data
# combined_data = { **processed_data_RS_baseline, **processed_data_RS_1H, **processed_data_RS_24H, **processed_data_RS_Day7 }
subjects_dict = create_subject_dictionaries(combined_data)

# Access the data for a specific subject, e.g., subject '004'
subject_004_data = subjects_dict.get('004', {})
print(subject_004_data.keys())  # Should print: dict_keys(['RS_baseline', 'RS_1H', 'RS_24H', 'RS_Day7'])

# Example: Accessing the 'RS_1H' data for subject '004'
subject_004_RS_1H_data = subject_004_data.get('RS_1H', None)
print(subject_004_RS_1H_data)  # Will print None if 'RS_1H' data is missing
```

```python
#Create a dictionary
import pandas as pd
import numpy as np

def compute_deltas(subjects_dict):
    """
    Compute delta values for each frequency band and timepoint compared to the baseline.
    Returns a dictionary where each key is a subject ID and the value is a DataFrame containing the deltas and boolean columns.
    """
    final_subjects_dict = {}
    
    # Define the timepoints and columns to calculate deltas for
    timepoints = ['RS_Day7', 'RS_24H', 'RS_1H']
    delta_columns = ['tc', 'dtc', 'o', 's']
    bands = ['Alpha', 'Beta', 'Theta', 'Gamma', 'Wholeband']
    
    for subject_id, subject_data in subjects_dict.items():
        # Create an empty DataFrame for the final output for this subject
        combined_df = pd.DataFrame()

        # Extract boolean rows (var_0 to var_14) from 'RS_baseline' and any frequency band
        bool_columns = [f'var_{i}' for i in range(15)]  # var_0 to var_14
        
        # Extract baseline data safely
        baseline_data = subject_data.get('RS_baseline', {})
        
        # Check if baseline_data is None and handle accordingly
        if baseline_data is None:
            baseline_data = {}
        
        alpha_data = baseline_data.get('Alpha', {})
        
        # Create DataFrame for boolean columns with data if available
        if alpha_data:
            bool_data = pd.DataFrame({
                var: alpha_data.get(var, [None] * len(alpha_data.get('tc', [])))
                for var in bool_columns
            })
        else:
            # Handle case where baseline data is not available
            bool_data = pd.DataFrame({var: [None] * 0 for var in bool_columns})
        
        # Add the boolean data to the combined DataFrame
        combined_df = pd.concat([combined_df, bool_data], axis=1)
        
        # Iterate over each frequency band
        for band in bands:
            # Extract baseline data for the current band
            band_baseline_data = baseline_data.get(band, {})

            # Check if band_baseline_data is None and handle accordingly
            if band_baseline_data is None:
                band_baseline_data = {}

            # Calculate deltas for each delta column
            for column in delta_columns:
                # Extract the baseline data for the column
                baseline_column_data = band_baseline_data.get(column, None)
                
                # Calculate deltas for each timepoint
                for timepoint in timepoints:
                    # Extract current timepoint data safely
                    timepoint_data = subject_data.get(timepoint, {})
                    
                    # Check if timepoint_data is None and handle accordingly
                    if timepoint_data is None:
                        timepoint_data = {}

                    band_timepoint_data = timepoint_data.get(band, {})

                    # Check if band_timepoint_data is None and handle accordingly
                    if band_timepoint_data is None:
                        band_timepoint_data = {}

                    current_column_data = band_timepoint_data.get(column, None)

                    # Calculate delta or set to None if data is missing
                    if baseline_column_data is not None and current_column_data is not None:
                        delta_values = current_column_data - baseline_column_data
                    else:
                        # Use the length of the baseline data series or default to an empty list
                        delta_length = len(baseline_column_data) if baseline_column_data is not None else len(alpha_data.get(column, []))
                        delta_values = pd.Series([None] * delta_length)

                    # Generate a descriptive column name for the delta
                    delta_column_name = f"{band.lower()}_{column}_delta_{timepoint.split('_')[1].lower()}"
                    combined_df[delta_column_name] = delta_values

        # Store the combined DataFrame for this subject in the final dictionary
        final_subjects_dict[subject_id] = combined_df

    return final_subjects_dict

# Compute deltas for all subjects
subjects_dict_all_deltas = compute_deltas(subjects_dict)
```
