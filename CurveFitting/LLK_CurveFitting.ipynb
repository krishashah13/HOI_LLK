```python
def quadratic_poly(x, a, b, c):
    return a * x**2 + b * x + c
```

```python
def fit_and_plot(df, measure, frequency_band):
    # Define timepoints mapping
    timepoints = {'1H': 1, '24H': 24, 'Day7': 168}  # 1 hour, 24 hours, and Day 7 (in hours)
    
    # Create a figure with two subplots, one for each group
    fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)
    
    # Loop over each group (0 and 1)
    for group, ax in zip([0, 1], axes):
        # Filter dataframe for the specific group, measure, and frequency band
        filtered_df = df[(df['measure'] == measure) & (df['frequency_band'] == frequency_band) & (df['Group'] == group)]
        
        # Loop over each subject in the group
        for subject_id in filtered_df['subject_id'].unique():
            subject_data = filtered_df[filtered_df['subject_id'] == subject_id]
            
            # Get x and y data
            x_data = np.array([timepoints[tp] for tp in ['1H', '24H', 'Day7']])
            y_data = subject_data[['1H', '24H', 'Day7']].values.flatten()
            
            # Perform quadratic polynomial fit (3 parameters: a, b, c)
            popt, _ = curve_fit(quadratic_poly, x_data, y_data)
            
            # Generate values for plotting the polynomial
            x_fit = np.linspace(1, 168, 100)  # From 1H to Day 7 (in hours)
            y_fit = quadratic_poly(x_fit, *popt)
            
            # Plot the fitted curve on the subplot for the current group
            ax.plot(x_fit, y_fit, label=f'Subject: {subject_id}')
            
            # Plot the original data points
            ax.scatter(x_data, y_data, color='black')  # Original points
        
        # Add title, labels, and grid to the subplot
        ax.set_title(f'Group {group} - {frequency_band} - Measure: {measure}')
        ax.set_xlabel('Time (hours)')
        ax.grid(True)
        ax.legend()
    
    # Add shared y-label and adjust layout
    fig.text(0.04, 0.5, f'{measure} value', va='center', rotation='vertical')
    plt.tight_layout()
    plt.show()
```

```python
# Example usage:
# Assuming you have the dataframe loaded, and you want to fit for 'O' and 'Alpha' band
fit_and_plot(features_data, measure='O', frequency_band='Alpha')
```
